# 1. Write a python program with exception handling to input marks for 5 subjects Physics, Chemistry, Biology,  Mathematics and Computer. Calculate the percentage and grade according to the following:
     
       i) Percentage>=90% : Grade A
      Ii) Percentage>=80% : Grade B
     iii) Percentage>=70% : Grade C
      iv) Percentage>=60% : Grade D
       v) Percentage>=40% : Grade E
      vi) Percentage<=40% : Grade F

```python
def grades(lst):
    percent=(sum(lst)/500)*100
    try:
        for i in lst:
            if i<0 or i>100:
                raise ValueError("Marks should not be Negative and Not more than 100 ")
                
        if percent>=90:
            print("Your Grade is A")
        elif percent>=80:
            print("Your Grade is B")
        elif percent>=70:
            print("Your Grade is C")
        elif percent>60:
            print("Your Grade is D")
        elif percent>40:
            print("Your Grade is E")
        else:
            print("Your Grade is F")
            
            
    except ValueError as e:
        print(e)
         
marks_list=[] 
phy=int(input("Enter Marks in Physics Subject : "))
marks_list.append(phy)
chem=int(input("Enter Marks in Chemistry Subject : "))
marks_list.append(chem)
bio=int(input("Enter Marks in Biology Subject : "))
marks_list.append(bio)
mat=int(input("Enter Marks in Mathematics Subject : "))
marks_list.append(mat)
comp=int(input("Enter Marks in Computer Subject : "))
marks_list.append(comp)

grades(marks_list)
```

# 2. write a python program with exception handling to input electricity unit charges and calculate the total electricity bill according to given Condition

     i) For the first 50 units Rs. 0.50/unit
    ii) For the next 100 units Rs. 0.75/unit
   iii) For the next 100 units Rs. 1.20/unit
    iv) For the units above 250 Rs. 1.50/unit
     v) An Additional Surcharge of 20% is added to bill:

```python
def bill_calculator(u):
    try:
        
        if u<0:
            raise ValueError("Units Should be Positive Integer value")
        if u<=50:
            bill=u*0.5
            surcharge=0.20*bill
            total_bill=bill+surcharge
            print("Your Electricity Bill amount = ",total_bill)
        elif u<=150:
            bill=50*0.50+(u-50)*0.75
            surcharge=0.20*bill
            total_bill=bill+surcharge
            print("Your Electricity Bill amount = ",total_bill)
        elif u<=250:
            bill=50*0.50+100*0.75+(u-150)*1.20
            surcharge=0.20*bill
            total_bill=bill+surcharge
            print("Your Electricity Bill amount = ",total_bill)
        else:
            bill=50*0.50+100*0.75+100*1.20+(u-250)*1.50
            surcharge=0.20*bill
            total_bill=bill+surcharge
            print("Your Electricity Bill amount = ",total_bill)
            
    except ValueError as e:
        print(e)
        
units=int(input("Enter No.of  Units = "))
bill_calculator(units)
```

# 3. Write a Python program with exception handling to input week number and weekday

```python
def week_day(n):
    try:
        
        if n>7 or n<0:
            raise ValueError("Week Number should be between 1 and 7")

        if n == 1 :
            print("Sunday")
        elif n == 2:
            print("Monday");

        elif n == 3:
            print("Tuesday")

        elif(n == 4) :
            print("Wednesday")

        elif(n == 5) :
            print("Thursday")

        elif(n == 6) :
            print("Friday")

        else:
            print("Saturday")
    except ValueError as e:
        print(e)
        


        
weekNumber= int(input("Enter the value of  week number : "))
week_day(weekNumber)
```

# 4. Write a python program to implement word using command line arguements

   i) Create a text document "apple.txt" which contains text for word count
  ii) Create a word count program which calls the "apple.txt" document by opening file
 iii) If the word is present again in the "apple.txt" , the wordcount is incremented by 1 until all the words are counted in          the document
  iv) Close the file
   v) Create a command.py program which imports the wordcount.py program
  vi) Count the number of words using command line arguments
 vii) Print each word and its count.

```python
# program for words count saved as wordcount.py
def words_count(f):
    with open(f,"r") as file:
        data=file.read()
        data_list=data.lower().split()
        data_list.sort()
        frequent_words_dict={}
        for i in data_list :
            if i.isalpha():
                count=data_list.count(i)
                frequent_words_dict[i]=count

    for j in frequent_words_dict:
        print(j,":",frequent_words_dict[j])
        
        
file_name="apple.txt"
words_count(file_name)
```

```python
# program for command.py
import sys
from wordcount import words_count
if len(sys.argv) != 2:
    print("Usage: python wordcount.py <filename>")
else:
    filename = sys.argv[1]
    w_count = words_count(filename)
    if w_count:
        for word, count in w_count.items():
            print(f"{word}: {count} times")
```

# 5. Write a python program for finding the most frequent words in a text read from a file

   i) Initially open the text file in read mode.
  ii) Make all the letters in the document into lowercase letters and split the each words in each line.
 iii) Get the words in a order.
  iv) Sort the words for finding the most frequent words in the file.
   v) Print the most frequent word in the file

```python
def count_words(f):
    with open(f,"r") as file:
        data=file.read()
        data_list=data.lower().split()
        frequent_words_dict={}
        for i in data_list :
            if i.isalpha():
                count=data_list.count(i)
                frequent_words_dict[i]=count
        words_sorted_list = sorted(frequent_words_dict.items(), key=lambda x: x[1],reverse=True)
        words_sorted=dict(words_sorted_list)
        #print(words_sorted)

    for i in words_sorted:
        print(i,":",words_sorted[i])
        
        
file_name="apple.txt"
count_words(file_name)
```
