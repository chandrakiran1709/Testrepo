# 2. Virtual environment management

```python
import subprocess
import sys
import os

def virtualenv_create(venv_name):
    try:
       
        subprocess.run(["python", "-m", "venv", venv_name])
        print(f"Virtual environment '{venv_name}' created successfully.")
    except Exception as e:
        print(f"Error creating virtual environment: {e}")

def virtualenv_activate(venv_name):
    try:
        if sys.platform == "win32":
            activate_script = os.path.join(venv_name, "Scripts", "activate")
        else:
            activate_script = os.path.join(venv_name, "bin", "activate")
        
        subprocess.run([activate_script], shell=True)
        print(f"Activated virtual environment '{venv_name}'.")
    except Exception as e:
        print(f"Error activating virtual environment: {e}")

def virtualenv_deactivate():
    try:
        
        subprocess.run(["deactivate"], shell=True)
        print("Virtual environment deactivated.")
    except Exception as e:
        print(f"Error deactivating virtual environment: {e}")

def package_install(venv_name, package_name):
    try:
      
        subprocess.run([os.path.join(venv_name, "bin", "pip"), "install", package_name])
        print(f"Package '{package_name}' installed in '{venv_name}'.")
    except Exception as e:
        print(f"Error installing package: {e}")

def package_upgrade(venv_name, package_name):
    try:
        subprocess.run([os.path.join(venv_name, "bin", "pip"), "install", "--upgrade", package_name])
        print(f"Package '{package_name}' upgraded in '{venv_name}'.")
    except Exception as e:
        print(f"Error upgrading package: {e}")

def package_uninstall(venv_name, package_name):
    try:
       
        subprocess.run([os.path.join(venv_name, "bin", "pip"), "uninstall", "-y", package_name])
        print(f"Package '{package_name}' uninstalled from '{venv_name}'.")
    except Exception as e:
        print(f"Error uninstalling package: {e}")

def packages_installed_list(venv_name):
    try:
        result = subprocess.run([os.path.join(venv_name, "bin", "pip"), "list"], capture_output=True, text=True)
        installed_packages = result.stdout
        print(f"Installed packages in '{venv_name}':\n{installed_packages}")
    except Exception as e:
        print(f"Error listing installed packages: {e}")

if _name_ == "__main__":
    venv_name = "Sample_venv"

    virtualenv_create(venv_name)
    virtualenv_activate(venv_name)

    package_install(venv_name, "requests")
    package_install(venv_name, "numpy")

    packages_installed_list(venv_name)

    package_upgrade(venv_name, "requests")

    packages_installed_list(venv_name)

    package_uninstall(venv_name, "numpy")

    deactivate_virtualenv()
```

# 3. Module Dependency Resolution

```python
import subprocess
import sys

def dependencies_install(requirements_file):
    try:
        subprocess.run(["pip", "install", "-r", requirements_file], check=True)
        print("Dependencies are successfully installed.")
    except subprocess.CalledProcessError as e:
        print(f"Error installing dependencies: {e}")
        sys.exit(1)


requirements = "require.txt"
dependencies_install(requirements)
```

## 2. Customer Order Processing

```python
import mysql.connector


def create_db_connection():
    try:
        connection = mysql.connector.connect(
            host="localhost",   
            user="root",  
            password="7989272689",  
            database="order_processing"
        )
        if connection.is_connected():
            return connection
    except Error as e:
        print(f"Error: {e}")
    return None

def place_order(connection, customer_id, product_id, quantity_ordered):
    try:
        cursor = connection.cursor()
        cursor.execute("SELECT quantity_available FROM Products WHERE product_id = %s", (product_id,))
        available_quantity = cursor.fetchone()[0]
        
        if available_quantity >= quantity_ordered:
            cursor.execute("SELECT price FROM Products WHERE product_id = %s", (product_id,))
            product_price = cursor.fetchone()[0]
            line_total = product_price * quantity_ordered
            cursor.execute("INSERT INTO Orders (customer_id, total_cost) VALUES (%s, %s)", (customer_id, line_total))
            order_id = cursor.lastrowid 
            cursor.execute("INSERT INTO OrderDetails (order_id, product_id, quantity_ordered, line_total) VALUES (%s, %s, %s, %s)",
                           (order_id, product_id, quantity_ordered, line_total))
            new_quantity = available_quantity - quantity_ordered
            cursor.execute("UPDATE Products SET quantity_available = %s WHERE product_id = %s", (new_quantity, product_id))
            
            connection.commit()
            print("Order placed successfully.")
        else:
            print("Product not available in sufficient quantity.")
    
    except Error as e:
        print(f"Error: {e}")


def generate_order_report(connection):
    try:
        cursor = connection.cursor()
        cursor.execute("SELECT Orders.order_id, Customers.first_name, Customers.last_name, Orders.order_date, Orders.total_cost "
                       "FROM Orders "
                       "INNER JOIN Customers ON Orders.customer_id = Customers.customer_id")
        order_data = cursor.fetchall()
        
        print("Order Report:")
        for order in order_data:
            print(f"Order ID: {order[0]}, Customer: {order[1]} {order[2]}, Order Date: {order[3]}, Total Cost: ${order[4]}")
    
    except Error as e:
        print(f"Error: {e}")

# Main function
def main():
    connection = create_db_connection()
    if connection:
        place_order(connection, 3, 12, 2)  
        generate_order_report(connection)   
        connection.close()

if _name_ == "__main__":
    main()
```

## 3. MySql Data base operations

```python
import mysql.connector


def create_db_connection():
    try:
        connection = mysql.connector.connect(
            host="localhost",     
            user="your_username",  
            password="7989272689",  
            database="mydatabase"  
        )
        if connection.is_connected():
            return connection
    except Error as e:
        print(f"Error: {e}")
    return None

def retrieve_data(connection):
    try:
        cursor = connection.cursor()
        cursor.execute("SELECT id, value FROM mytable")
        data = cursor.fetchall()
        return data
    except Error as e:
        print(f"Error: {e}")
        return []


def perform_operations(data):
    modified_data = []
    for row in data:
        id, value = row
        modified_value = value * 2
        modified_data.append((id, modified_value))
    return modified_data


def update_database(connection, modified_data):
    try:
        cursor = connection.cursor()
        for id, modified_value in modified_data:
            cursor.execute("UPDATE mytable SET value = %s WHERE id = %s", (modified_value, id))
        connection.commit()
        print("Data updated successfully.")
    except Error as e:
        connection.rollback()
        print(f"Error: {e}")


def main():
    connection = create_db_connection()
    if connection:
        data = retrieve_data(connection)
        if data:
            modified_data = perform_operations(data)
            update_database(connection, modified_data)
            connection.close()

if _name_ == "__main__":
    main()
```
