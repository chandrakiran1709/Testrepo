```python
%matplotlib inline
import pandas as pd
```

```python
cast = pd.read_csv('cast.csv', index_col=None)
cast.head()
```

```python
release_dates = pd.read_csv('release_dates.csv', index_col=2, parse_dates=['date'])
release_dates.head()
```

## 1.Make a bar plot of the months in which movies with "Christmas" in their title tend to be released in the USA.

```python
import matplotlib.pyplot as pt
mt = release_dates[(release_dates.index == 'USA') & (release_dates['title'].str.contains('Christmas'))]
mt.date.dt.month.value_counts().sort_index().plot(kind='bar')
pt.xlabel('Month')
pt.ylabel('No.of Movies')
```

## 2.Make a bar plot of the months in which movies whose titles start with "The Hobbit" are released in the USA.

```python
mt = release_dates[
    (release_dates['title'].str.startswith('The Hobbit')) &
    (release_dates.index == 'USA')
]
mt.date.dt.month.value_counts().sort_index().plot(kind='bar')
pt.title('Months of Movie Releases with "The Hobbit" in Title (USA)')
pt.xlabel('Month')
pt.ylabel('Number of Movies')
```

## 3. Make a bar plot of the day of the week on which movies with "Romance" in their title tend to be released in the USA.

```python
mt = release_dates[
    (release_dates['title'].str.contains('Romance')) &
    (release_dates.index == 'USA')
]

mt.date.dt.dayofweek.value_counts().sort_index().plot(kind='bar')
pt.title('Day of the Week for Movie Releases with "Romance" in Title (USA)')
pt.xlabel('Day of the Week')
pt.ylabel('Number of Movies')
```

## 4.Make a bar plot of the day of the week on which movies with "Action" in their title tend to be released in the USA.

```python
mt= release_dates[
    (release_dates['title'].str.contains('Action')) &
    (release_dates.index == 'USA')
]
mt.date.dt.dayofweek.value_counts().sort_index().plot(kind='bar')
pt.title('Day of the Week for Movie Releases with "Action" in Title (USA)')
pt.xlabel('Day of the Week')
pt.ylabel('Number of Movies')
```

## 5.On which date was each Judi Dench movie from the 1990s released in the USA?

```python
usa = release_dates[release_dates.index == 'USA']
c = cast
c = c[c.name == 'Judi Dench']
c = c[c.year // 10 * 10 == 1990]
c.merge(usa).sort_values('date')
```

## 6.In which months do films with Judi Dench tend to be released in the USA

```python
c = cast
c = c[c.name == 'Judi Dench']
mt = c.merge(usa).sort_values('date')
mt.date.dt.month.value_counts().sort_index().plot(kind='bar')
pt.title('Months of Movie Releases Featuring Judi Dench in the USA')
pt.xlabel('Month')
pt.ylabel('Number of Movies')
```
